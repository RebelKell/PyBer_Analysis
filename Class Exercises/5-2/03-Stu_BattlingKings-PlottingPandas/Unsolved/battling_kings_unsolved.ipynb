```python
# Dependencies
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

```python
# Read CSV
data_to_load = "Resources/got.csv"
```

```python
pd.read_csv(data_to_load)
got_df = pd.read_csv(data_to_load)
got_df.head()
```

```python
# Get attacker and defender king data
attacker_data = got_df["attacker_king"].value_counts()
defender_data = got_df["defender_king"].value_counts()
attacker_data
```

```python
# Get total battle data. Fill value replaces null values with "0"
battle_data = attacker_data.add(defender_data, fill_value=0)
battle_data
```

```python
# Configure plot and ticks
battle_data.plot(kind="bar", facecolor="blue")

# Set textual properties


# Show plot


# Resize plot to display labels
```
