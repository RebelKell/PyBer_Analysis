```python
# Import dependencies.
import matplotlib.pyplot as plt
```

```python
## CREATING LINE CHARTS
```

```python
# Set the x-axis to a list of strings for each month.
x_axis = ["Jan", "Feb", "Mar", "April", "May", "June", "July", "Aug", "Sept", "Oct", "Nov", "Dec"]

# Set the y-axis to a list of floats as the total fare in US dollars accumulated for each month.
y_axis = [10.02, 23.24, 39.20, 35.42, 32.34, 27.04, 43.82, 10.56, 11.85, 27.90, 20.71, 20.09]
```

```python
# Create the plot using MATLAB Method
plt.plot(x_axis, y_axis)
```

```python
# Create the plot using object-oriented interface with ax.plt()
fig, ax = plt.subplots()
ax.plot(x_axis, y_axis)
```

```python
# Create the plot using object-oriented interface with ax.plt(): ALTERNATE METHOD
fig = plt.figure()
ax = fig.add_subplot()
ax.plot(x_axis, y_axis)
```

```python
# Create the plot using object-oriented interface with ax.plt(): ANOTHER, FASTER METHOD!
ax = plt.axes()
ax.plot(x_axis, y_axis)
```

```python
# Create the plot and add a label for the legend using MATLAB
plt.plot(x_axis, y_axis, marker="*", color="blue", linewidth=2, label='Boston')
# Create labels for the x and y axes.
plt.xlabel("Date")
plt.ylabel("Fare($)")
# Set the y limit between 0 and 45.
plt.ylim(0, 45)
# Create a title.
plt.title("PyBer Fare by Month")
# Add a grid.
plt.grid()
# Add the legend.
plt.legend()
# set an x-axis limit: plt.xlim("Jan", "June") or using indices: plt.xlim(0, 5)
```

```python
# Create the plot and add a label for the legend using Object-Oriented interface
ax = plt.axes()
ax.plot(x_axis, y_axis, color="green", marker="*", linewidth=2, label="Boston")
ax.set_title("Pyber Fare by Month")
ax.set_xlabel('Date')
ax.set_ylabel('Fare($)')
ax.legend()
ax.grid()
```

```python
## CREATING BAR CHARTS USING MATLAB
```

```python
# Set the x-axis to a list of strings for each month.
x_axis = ["Jan", "Feb", "Mar", "April", "May", "June", "July", "Aug", "Sept", "Oct", "Nov", "Dec"]

# Set the y-axis to a list of floats as the total fare in US dollars accumulated for each month.
y_axis = [10.02, 23.24, 39.20, 35.42, 32.34, 27.04, 43.82, 10.56, 11.85, 27.90, 20.71, 20.09]
```

```python
# Create a Vertical Bar Chart using the MATLAB method
plt.bar(x_axis, y_axis)
```

```python
# Create the plot.
plt.bar(x_axis, y_axis, color="green", label='Boston')
# Create labels for the x and y axes.
plt.xlabel("Date")
plt.ylabel("Fare($)")
# Create a title.
plt.title("PyBer Fare by Month")
# Add the legend.
plt.legend()
```

```python
# Create a Horizontal Bar Chart using the MATLAB method
plt.barh(x_axis, y_axis, color="magenta", label = "Boston")
#invert x and y axis
plt.gca().invert_yaxis()
# Format the Horizontal Bar Chart 
plt.title("PyBer Fare by Month")
plt.xlabel("Fare($)")
plt.ylabel("Date")
plt.legend()
```

```python
# Set the x-axis to a list of strings for each month.
x_axis = ["Jan", "Feb", "Mar", "April", "May", "June", "July", "Aug", "Sept", "Oct", "Nov", "Dec"]

# Set the y-axis to a list of floats as the total fare in US dollars accumulated for each month.
y_axis = [10.02, 23.24, 39.20, 35.42, 32.34, 27.04, 43.82, 10.56, 11.85, 27.90, 20.71, 20.09]
```

```python
# Create a vertical bar chart with ax.plt()
fig, ax = plt.subplots(figsize=(8,8))
ax.bar(x_axis, y_axis, color="black", label="New York")
ax.set_title("Pyber Fare by Month")
ax.set_xlabel('Date')
ax.set_ylabel('Fare($)')
ax.legend()
```

```python
# Create horizontal bart chart with ax.plt()
fig, ax = plt.subplots()
ax.barh(x_axis, y_axis, color="cyan", label="Chicago")
#invert x and y axis
plt.gca().invert_yaxis()
ax.set_title("Pyber Fare by Month")
ax.set_xlabel('Fare($)')
ax.set_ylabel('Date')
ax.legend()
```

```python
ax.plot(x_axis, y_axis, color="green", marker="*", linewidth=2, label="Boston")
ax.set_title("Pyber Fare by Month")
ax.set_xlabel('Date')
ax.set_ylabel('Fare($)')
ax.legend()
ax.grid()
```

```python
#Create a Pie Chart Using the MATLAB Method
```

```python
# Assign 12 colors, one for each month.
colors = ["slateblue", "magenta", "lightblue", "green", "yellowgreen", "greenyellow", "yellow", "orange", "gold", "indianred", "tomato", "mistyrose"]
#Explode the July value
explode_values = (0, 0, 0, 0, 0, 0, 0.2, 0, 0, 0, 0, 0)
# Increase spacing of percentages
plt.subplots(figsize=(8, 8))
plt.pie(y_axis,
    explode=explode_values,
    colors=colors,
    labels=x_axis,
    autopct='%.1f%%')

plt.show()
```

```python
# Create a Pie Chart Using the Object-Oriented Interface
```

```python
# SKILL DRILL 5.1.8: Using the object-oriented approach, make the following changes to the pie chart:
colors = ["slateblue", "magenta", "lightblue", "green", "yellowgreen", "greenyellow", "yellow", "orange", "gold", "indianred", "tomato", "mistyrose"]
explode_values = (0, 0, .2, 0, 0, 0, 0.2, 0, 0, 0, 0, 0)

fig, ax = plt.subplots(figsize=(8, 8))
ax.pie(y_axis,
       labels=x_axis,
       shadow=True,
       explode=explode_values,
       startangle=85,
       autopct='%.1f%%',
       colors = colors
        )

plt.show()
```
